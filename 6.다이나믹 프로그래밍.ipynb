### 다이나믹 프로그래밍 : 메모리를 적절히 사용하여 수행 시간 효율성을 비약적으로 향상시키는 방법

- 이미 계산된 결과(작은 문제)는 별도의 메모리 영역에 저장하여 다시 계산하지 않도록 하기 때문에 시간 복잡도를 줄일 수 있어 효율적. 다이나믹 프로그래밍의 '다이나믹' 은 동적할당의 의미와 다름

구현방식은 탑다운과 바텀업 방식으로 구성됩니다.

다음 조건을 만족할 때 사용할 수 있습니다.
1. 최적 부분 구조(Optimal Substructure)
        큰 문제를 작은 문제로 나눌 수 있으며 작은 문제의 답을 모아서 큰 문제를 해결할 수 있습니다.
2. 중복되는 부분 문제(Overlapping Subproblem)
        동일한 작은 문제를 반복적으로 해결해야 합니다

```python
# 피보나치 함수(Fibonacci Fucntion)을 재귀함수로 구현
def fibo(x):
    # 피보나치 함수 a1 = 1, a2 = 1 이므로 return 1 값을 호출하여, 정의내림
    if x == 1 or x == 2: 
        return 1
    return fibo(x - 1) + fibo(x - 2)

# 피보나치 함수는 지수 시간 복잡도를 가지게 되는데 
# 중복되는 부분 문제는 별도의 메모리 공간에 저장하여 효율적으로 계산하게 바꾼다

''' 계산속에 f(n-x) 가 계속 중복되는 것을 확인 할 수 있다
 f(6) = f(5) + f(4) = {f(4) + f(3)} + {f(3) + f(2)} 
      = [{f(3) + f(2)} + {f(2) + f(1)}] + [{f(2) + f(1)} + f(2)]
'''

fibo(4)
```

- 메모이제이션(Memoization) : 탑다운 방식(하향식), 
한번 계산한 결과를 메모리 공간에 메모
값을 기록해 놓는다는 점에서 캐싱(Caching)이라고도 함

- 메모이제이션은 한번 계산된 결과를 일시적으로 기록해 놓는 넓은 개념을 의마하며, 다이나믹 프로그래밍을 위해 활용되지 않을 수도 있어, 엄밀히 말하면 서로 다른 개념


다이나믹 프로그래밍의 전형적인 형태는 바텀업 방식.

- 결과 저장용 리스트는 DP 테이블이라고 함

```python
# 한 번 계산된 결과를 메모이제이션(Memoization)하기 위한 리스트 초기화
'f(99)을 구할 때'
d = [0] * 100

# 피보나치 함수를 탑다운 다이나믹 프로그래밍 형태로 구현
def fibo(x):
    if x == 1 or x == 2:
        return 1 # 처음시작 a1,a2 지정
    
    # 이미 계산한 적 있는 문제라면 그대로 반환
    if d[x] != 0:
        return d[x] # 피보나치 함수를 통해 d[x] 값이 입력되면 리스트에 저장
    
    d[x] = fibo(x - 1) + fibo(x - 2)
    return d[x]

print(fibo(99))


# 바텀업 다이나믹 프로그래밍 방식
d = [0] * 100
d[1] = 1
d[2] = 1
n = 99
for i in range(3, n+1):
    d[i] = d[i - 1] + d[i -2]

print(d[n])
```

이 문제의 기본 아이디어는 <b>가장 긴 증가하는 부분 수열(Longest Increasing Susequence, LIS) </b> 로 알려진 전형적인 다이나믹 프로그래밍 문제의 아이디어와 같습니다

위 문제는 가장 긴 감소하는 부분 수열을 찾는 문제로 치환할 수 있으므로, LIS 알고리즘을 조금 수정하여 적용함으로써 정답을 도출할 수 있습니다.

```python
'''
가장 긴 증가하는 부분 수열 (LIS) 알고리즘
D[i] = array[i]를 마지막 원소로 가지는 부분 수열의 최대 길이

점화식
모든 0 <= j < i 에 대하여,
D[i] = max(D[i], D[j] + 1) if array[j] < array [i]
'''

n = int(input())
array = list(map(int, input().split()))
# 순서를 뒤집어 가장 증가하는 부분 수열 문제로 변환
array.reserse()

# 1차원 DP 테이블 초기화
dp = [1]* n # 모든 수가 자기 자신의 수로 1개 초기화

# LIS 알고리즘
for i in range(1,n):
    for j in range(0, 1):
        if array[j] < array[i]:
            dp[i] = max(dp[i], dp[j] + 1)
            
# 열외해야 하는 병사의 최소 수를 출력 = 즉 dp에 포함안된 값
print(n - max(dp))
```
